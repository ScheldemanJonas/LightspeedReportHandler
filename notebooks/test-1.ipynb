```python
import xml.etree.ElementTree as ET
import pandas as pd
from pathlib import Path
```

# Path to data

```python
ROOT = Path().absolute()
DATA = ROOT.parent / 'src' / 'data'
```

# Access data

```python
report = DATA / 'dayreport-2022-08-10.xml'
tree = ET.parse(report)
root = tree.getroot()
```

## Tax-types

```python
tax_types = []

for child in root.iter('Tax'):
    tax_types.append(child.get('rate'))

tax_types = set(tax_types)
tax_types
```

## Payment-types


```python
payment_types = []

for child in root.iter('Payment'):
    payment_types.append(child.get('type'))

payment_types = set(payment_types)
payment_types
```

## Iterating on Receipts

```python
d = []
for receipt in root.iter('Receipt'):
    print('================')

    payment_t = next(receipt.iter('Payment'))
    print('Payment - Type: ', payment_t.get('type'))

    for t in receipt.iter('Tax'):

        print('Tax - rate: ', t.get('rate'))
        print('Tax - amount: ', t.get('revenueTotal'))

        d.append(
            {
                'PaymentType': payment_t.get('type'),
                'Tax': t.get('rate'),
                'Amount':  float(t.get('revenueTotal'))
            }
        )

df = pd.DataFrame(d)
```
